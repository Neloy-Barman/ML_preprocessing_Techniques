## **`Column normalization`**

```python
def column_normalization(frame, column_name):
    """
    It takes the frame and column_name as input and returns us with the normalized column 
    placed in the frame.
    """
    if frame[column_name].min() >= 0.00 and frame[column_name].max() <= 1.00:
        return frame
    else:
        frame[column_name] = (frame[column_name] - frame[column_name].mean()) / frame[column_name].std()
        frame[column_name] = (frame[column_name] - frame[column_name].min()) / (frame[column_name].max() - frame[column_name].min())
        return frame
```

## **`Normalizing the frame`**

```python
def normalize_frame(frame, column_list):
    """
    It takes the frame and column_list as input and normalizes the column values using 
    column_normalization function.
    
    """
    for i in range(len(column_list)):
        frame = column_normalization(frame, column_list[i])
    return frame
```

```python
# The column index contains prediction class as that's not necessary so, it's dropped. 
data = data.drop(["Class"], axis=1)
Column_list = data.columns
print(f"{Column_list}")
```

```python
Normalized_frame = normalize_frame(data, Column_list)
```

```python
Normalized_frame.to_csv("N_Music_Emotion.csv", index= False)
```
